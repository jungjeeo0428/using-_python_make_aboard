```python
# CLI게시판
print("CLI게시판에 오신걸 환영합니다")
k = 0
article_number = []
article_title_list = []
text_title_list = []
while True:
  cmd = input("명령어:")
  if cmd == "exit":
    break
  elif cmd == "write article":
   line = input("제목:")
   article_title_list.append(line)
   line_text = input("내용:")
   text_title_list.append(line_text)
   k += 1
   article_number.append(k)
   print("{}번글 생성".format(k))

  if cmd == "article list":
    print("현재 게시글번호{} 타이틀{}내용{}".format(article_number, article_title_list, text_title_list))



print("==CLI 게시판 종료 ==")
```
